```python
out_data = pd.read_csv('./pol_t=60.csv', skiprows=None)
out_data
```

### First interpolate Ux

```python
len(out_data['U:0'])
```

```python
save_file = './U0.xy'
u0_arr = out_data['U:0']
vert_arr = out_data['Points:1']
# the overwrite mode is used here
with open(save_file, "w") as f:
    f.write("("+'\n')
    for item in range(len(u0_arr)):
        f.write("("+str(vert_arr[item])+" "+str(u0_arr[item])+")"+'\n')
    f.write(")")
    f.close()
```
